```python
iris = pd.read_excel("iris_xlsx.xlsx")
iris.head()
```

```python
iris_sub_t = iris_sub.transpose()
iris_sub_t
```

```python
iris_sub_t["diff_sq"] = (iris_sub_t[0] - iris_sub_t[1]) ** 2
iris_sub_t
```

```python
sl_min = iris["Sepal.Length"].min()
sl_max = iris["Sepal.Length"].max()
iris["SL_MM"] = (iris["Sepal.Length"] - sl_min) / (sl_max - sl_min)
iris.head()
```

```python
pl_mean = iris["Petal.Length"].mean()
pl_std = iris["Petal.Length"].std()
iris["PL_std"] = (iris["Petal.Length"] - pl_mean) / pl_std
print(iris["PL_std"].mean())
print(iris["PL_std"].std())
```

```python
def nor_mm(x):
    x_min = x.min()
    x_max = x.max()
    ret = (x - x_min) / (x_max - x_min)
    return ret
    
```

```python
nor_mm(pd.Series([1,2,3,4,5]))
```

```python
def nor_std(x):
    x_mean = x.mean()
    x_std = x.std()
    ret = (x - x_mean) / x_std
    return ret
```

```python
nor_std(pd.Series([1,2,3,4,5]))
```

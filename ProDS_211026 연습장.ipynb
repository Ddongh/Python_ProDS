```python
elec = pd.read_csv("elec_load.csv")
elec.head()
```

```python
elec.loc[:2, "X7HR" : "X15HR"]
```

```python
elec.iloc[0, ].reset_index()
```

```python
elec_melt = elec.melt(id_vars = ["YEAR", "MONTH", "DAY"])
elec_melt
```

```python
elec_melt = elec_melt.rename(columns = {"variable" : "HOUR", "value" : "LOAD"})
elec_melt.head()
```

```python
elec_melt["HOUR"] = elec_melt["HOUR"].str.replace(pat = "X", repl = "")
```

```python
elec_melt.head()
```

```python
elec_cast = elec_melt.pivot(index = ["YEAR", "MONTH", "DAY"], columns = ["HOUR"], values = ["LOAD"]).reset_index()
```

```python
elec_cast
```

```python
elec_cast = elec_melt.pivot_table(index = ["YEAR", "MONTH", "DAY"], columns = ["HOUR"], values = ["LOAD"]).reset_index()
```
